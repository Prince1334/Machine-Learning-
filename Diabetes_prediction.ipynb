```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import torch
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
df= pd.read_csv("diabetes_dataset.csv")
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum().sum()
```

```python
df.gender.unique()
```

```python
dumm = pd.get_dummies(df['gender'])
dumm
```

```python
df2 = pd.concat([df,dumm],axis="columns")
```

```python
df2.head(5)
```

```python
x = df2.drop(['age','gender','ethnicity','education_level','income_level','employment_status','diabetes_stage','smoking_status','diagnosed_diabetes','Female'],axis="columns")
y = df2.diagnosed_diabetes
```

```python
 X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(random_state= 42)
model.fit(X_train,y_train)
model.score(X_test,y_test)*100
```
